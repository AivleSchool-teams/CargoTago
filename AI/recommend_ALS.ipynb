```python
# 고유한 유저, 아티스트를 찾아내는 코드, indexing하기에 좋은 함수 unique()
user_unique = data['num'].unique()
dep_unique = data['Departure (3)'].unique()

# 유저, 아티스트 indexing 하는 코드 idx는 index의 약자입니다.
user_to_idx = {v:k for k,v in enumerate(user_unique)}
dep_to_idx = {v:k for k,v in enumerate(dep_unique)}

# indexing을 통해 데이터 컬럼 내 값을 바꾸는 코드
# user_to_idx.get을 통해 user_id 컬럼의 모든 값을 인덱싱한 Series를 구해 봅시다.
# 혹시 정상적으로 인덱싱되지 않은 row가 있다면 인덱스가 NaN이 될 테니 dropna()로 제거합니다.
temp_user_data = data['num'].map(user_to_idx.get).dropna()
if len(temp_user_data) == len(data):   # 모든 row가 정상적으로 인덱싱되었다면
    print('user_id column indexing OK!!')
    data['user_id'] = temp_user_data   # data['user_id']을 인덱싱된 Series로 교체해 줍니다.
else:
    print('user_id column indexing Fail!!')

# artist_to_idx을 통해 artist 컬럼도 동일한 방식으로 인덱싱해 줍니다.
temp_dep_data = data['Departure (3)'].map(dep_to_idx.get).dropna()
if len(temp_dep_data) == len(data):
    print('Departure column indexing OK!!')
    data['dep_id'] = temp_dep_data
else:
    print('Departure column indexing Fail!!')

data
```

# Hyperparameter

```python
from implicit.evaluation import precision_at_k, train_test_split
from implicit.als import AlternatingLeastSquares

# 데이터를 학습 세트와 테스트 세트로 분할
train, test = train_test_split(csr_data, train_percentage=0.8)

# Implicit AlternatingLeastSquares 모델의 선언
# 1. factors : 유저와 아이템의 벡터를 몇 차원으로 할 것인지
# 2. regularization : 과적합을 방지하기 위해 정규화 값을 얼마나 사용할 것인지
# 3. use_gpu : GPU를 사용할 것인지
# 4. iterations : epochs와 같은 의미입니다.

# 하이퍼파라미터 그리드
# factors : 잠재요인개수, regularization : 사용자-잠재요인, 아이템-잠재요인 행렬 최적화시 사용 , iteration : 최적화 반복횟
param_grid = {'factors': [10, 20, 50], 'regularization': [0.01, 0.1, 1], 'iterations': [10, 20, 50]}

# 각 하이퍼파라미터 조합에 대해 모델 학습 및 평가
best_score = 0
best_params = None

for factors in param_grid['factors']:
    for reg in param_grid['regularization']:
        for iterations in param_grid['iterations']:
            model = AlternatingLeastSquares(factors=factors, regularization=reg, iterations=iterations)
            model.fit(train)

            score = precision_at_k(model, train, test, K=10, num_threads=4)

            if score > best_score:
                best_score = score
                best_params = (factors, reg, iterations)

print(best_params)
```
