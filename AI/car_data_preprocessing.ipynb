```python
start = data[['NO.', 'code3', 'code4', 'code5', 'A2_1', 'A2_2', 'A2_4', 'A2_4_t', 'A2_5', 'A2_6', 'A7_1']]
start
```

#### 출발지, 도착지, 도착지유형

 - 출발지, 도착지, 도착지유형 받아오기
  - 칼럼 명 바꿔주기
  - 출발지 : Bn_8_1
  - 도착지 : Bn+1_8_1
  - 도착지유형 : Bn+1_9


```python
tmp = data[['A8_1_1',
'A8_1_2',
'A8_1_3',
'B1_8_1',
'B1_8_2',
'B1_8_3',
'B1_9',]]

tmp
```

```python
seed = pd.concat([start, tmp], axis = 1)

seed.rename(columns = {'A8_1_1' : 'Departure (1)',
'A8_1_2' : 'Departure (2)',
'A8_1_3' : 'Departure (3)',
'B1_8_1' : 'Destination (1)',
'B1_8_2' : 'Destination (2)',
'B1_8_3' : 'Destination (3)',
'B1_9' : 'Destination type'
}, inplace = True)

seed
```

```python
for i in range(1, 15):
    tmp = data[['B' + str(i) + '_8_1',
    'B' + str(i) + '_8_2',
    'B' + str(i) + '_8_3',
    'B' + str(i + 1) + '_8_1',
    'B' + str(i + 1) + '_8_2',
    'B' + str(i + 1) + '_8_3',
    'B' + str(i + 1) + '_9']]
    tmp_2 = pd.concat([start, tmp], axis = 1)
    
    tmp_2.rename(columns = {'B' + str(i) + '_8_1' : 'Departure (1)',
    'B' + str(i) + '_8_2' : 'Departure (2)',
    'B' + str(i) + '_8_3' : 'Departure (3)',
    'B' + str(i + 1) + '_8_1' : 'Destination (1)',
    'B' + str(i + 1) + '_8_2' : 'Destination (2)',
    'B' + str(i + 1) + '_8_3' : 'Destination (3)',
    'B' + str(i + 1) + '_9' : 'Destination type'
    }, inplace = True)
    final = pd.concat([seed, tmp_2], axis = 0)
    
final
```

```python
final['A2_4_t'].fillna(0, inplace = True)
```
